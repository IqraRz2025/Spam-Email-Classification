```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the dataset
df = pd.read_csv('emails.csv')

# Identify categorical columns (assuming object or category dtype)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

# Initialize the OneHotEncoder
# handle_unknown='ignore' is useful to prevent errors if new categories appear later
one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False) # Use sparse_output=False to get a dense array

# Apply one-hot encoding to the categorical columns
# You can fit_transform on a subset of your data or the relevant columns
encoded_data = one_hot_encoder.fit_transform(df[categorical_cols])

# Create a new DataFrame from the encoded data
# You would need to get the feature names after encoding for clarity
# For simplicity, we'll just create a DataFrame with generic column names
encoded_df = pd.DataFrame(encoded_data, columns=one_hot_encoder.get_feature_names_out(categorical_cols))

# Concatenate the encoded DataFrame with the original numerical columns (if any)
# This assumes you want to keep numerical columns as they are
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
df_processed = pd.concat([df[numerical_cols], encoded_df], axis=1)

# Display the first few rows of the processed DataFrame
print(df_processed.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv('emails.csv')

# Identify numerical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply scaling to the numerical columns
# The scaler expects a 2D array, so we select the numerical columns and use .values
scaled_data = scaler.fit_transform(df[numerical_cols])

# Create a new DataFrame from the scaled data
scaled_df = pd.DataFrame(scaled_data, columns=numerical_cols)

# Concatenate the scaled DataFrame with the original non-numerical columns (if any)
# This assumes you want to keep non-numerical columns as they are
non_numerical_cols = df.select_dtypes(exclude=['int64', 'float64']).columns
df_processed = pd.concat([df[non_numerical_cols], scaled_df], axis=1)

# Display the first few rows of the processed DataFrame
print(df_processed.head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Load the preprocessed dataset
# Assuming 'df_processed' from the previous preprocessing steps is available
# If not, replace this with loading and preprocessing your data
try:
    df_processed
except NameError:
    print("Please run the preprocessing code first to create 'df_processed'.")
    # Load original data and perform basic preprocessing for demonstration
    df = pd.read_csv('emails.csv')
    # Drop non-numeric columns for simplicity in this example
    # Assuming the 'text' or similar column is the one causing issues, drop it
    if 'text' in df.columns:
        df = df.drop(columns=['text'])
    df_processed = df.select_dtypes(include=['int64', 'float64']).dropna()
    # Assuming the last column is the target
    X = df_processed.drop(columns=[df_processed.columns[-1]])
    y = df_processed[df_processed.columns[-1]]


# Assuming 'y' is your target variable and 'X' contains your features
# If you preprocessed X and y separately, use those
# For demonstration, we'll use the simplified df_processed as X and the last column as y
if 'X' not in locals() or 'y' not in locals():
    # Ensure df_processed only contains numerical columns before splitting
    df_processed_numerical = df_processed.select_dtypes(include=['int64', 'float64'])

    # Assuming the last column is the target and all others are features
    # Check if df_processed_numerical is empty or has less than 2 columns
    if df_processed_numerical.empty or df_processed_numerical.shape[1] < 2:
         raise ValueError("df_processed does not contain enough numerical columns for features and target.")

    X = df_processed_numerical.drop(columns=[df_processed_numerical.columns[-1]])
    y = df_processed_numerical[df_processed_numerical.columns[-1]]


# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Logistic Regression ---
print("--- Logistic Regression ---")
# Initialize and train the model
logreg = LogisticRegression(max_iter=1000) # Increase max_iter if convergence warning
logreg.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_logreg = logreg.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("\nClassification Report:\n", classification_report(y_test, y_pred_logreg))

# --- Cross-validation for Logistic Regression ---
print("\n--- Cross-validation (Logistic Regression) ---")
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation
cv_scores = cross_val_score(logreg, X, y, cv=5)

print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
```

the code for Random Forest and Gradient Boosting:

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming X and y are already defined from the previous preprocessing steps
# If not, load and preprocess your data as before to get numerical X and y

# --- Random Forest ---
print("\n--- Random Forest ---")
# Initialize and train the model
# n_estimators: the number of trees in the forest
# random_state: for reproducibility
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nClassification Report:\n", classification_report(y_test, y_pred_rf))

# --- Cross-validation for Random Forest ---
print("\n--- Cross-validation (Random Forest) ---")
cv_scores_rf = cross_val_score(rf_model, X, y, cv=5)
print("Cross-validation scores:", cv_scores_rf)
print("Average cross-validation score:", cv_scores_rf.mean())

# --- Gradient Boosting ---
print("\n--- Gradient Boosting ---")

gb_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
gb_model.fit(X_train, y_train)

# Make predictions
y_pred_gb = gb_model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred_gb))
print("\nClassification Report:\n", classification_report(y_test, y_pred_gb))

# --- Cross-validation for Gradient Boosting ---
print("\n--- Cross-validation (Gradient Boosting) ---")
cv_scores_gb = cross_val_score(gb_model, X, y, cv=5)
print("Cross-validation scores:", cv_scores_gb)
print("Average cross-validation score:", cv_scores_gb.mean())
```
